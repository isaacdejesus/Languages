a_list[m:n] allows us to slice a list. m is first element and n + 1 for the last element

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
row_2 = ['Instagram', 0.0, 'USD', 2161558, 4.5]
row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]
row_4 = ['Temple Run', 0.0, 'USD', 1724546, 4.5]
row_5 = ['Pandora - Music & Radio', 0.0, 'USD', 1126879, 4.0]


first_4_fb = row_1[:4]
last_3_fb = row_1[-3:]
pandora_3_4 = row_5[2:4]
print(first_4_fb, last_3_fb, pandora_3_4)

```

use loop to calculate ratings

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
row_2 = ['Instagram', 0.0, 'USD', 2161558, 4.5]
row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]
row_4 = ['Temple Run', 0.0, 'USD', 1724546, 4.5]
row_5 = ['Pandora - Music & Radio', 0.0, 'USD', 1126879, 4.0]

app_data_set = [row_1, row_2, row_3, row_4, row_5]

rating_sum = 0
for elements in app_data_set:
    rating = elements[-1]
    rating_sum += rating
avg_rating = rating_sum/5
```

```python
opened_file = open('AppleStore.csv')
from csv import reader
read_file = reader(opened_file)
apps_data = list(read_file)

rating_sum = 0
#start with [1:] to exclude the header row
for row in apps_data[1:]:
    #convert strings to floats
    rating = float(row[7])
    rating_sum+=rating
avg_rating = rating_sum/7197
```

```python
opened_file = open('AppleStore.csv')
from csv import reader
read_file = reader(opened_file)
apps_data = list(read_file)

#create empty list to hold all the ratings
all_ratings = []
for row in apps_data[1:]:
    #obtain rating from each row
    rating = float(row[7])
    #add each rating to all_ratings
    all_ratings.append(rating)
#use sum function to add up all ratings
#use len to find lenght of list
avg_rating = sum(all_ratings)/len(all_ratings)
print(avg_rating)
```
